# Linear Regression

## Init datapoints

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear data
k = 3 # slope
m = 2 # intersection
perturbation = 10
x = np.linspace(-3, 3)
noise = np.random.random(x.shape) - 0.5
y = k * x + m + noise * perturbation

plt.scatter(x, y)
plt.show()
```

## ML Estimator

```python
from numpy.linalg import inv
# Assume data follows: y = X * theta + error
# X: design matrix
X = np.array([np.ones_like(x), x]).T

theta_ML = inv(X.T @ X) @ X.T @ y

y_hat = theta_ML @ X.T

plt.scatter(x, y, label=f"points following y = {k}x + {m} + error")
plt.plot(x, y_hat, label=fr"ML estimator: $\theta_1={theta_ML[1]:.3f}$, $\theta_0={theta_ML[0]:.3f}$")
plt.legend()
plt.show()
```

## Quadradic data

```python
# Quadratic
a = 3
b = 2
c = 4
perturbation = 10
x = np.linspace(-3, 3)
noise = np.random.random(x.shape) - 0.5
y = a * x**2 + b * x + c + noise * perturbation

plt.scatter(x, y)
plt.show()
```

```python
X = np.array([np.ones_like(x), x, x**2]).T

theta_ML = inv(X.T @ X) @ X.T @ y

y_hat = theta_ML @ X.T

plt.scatter(x, y, label=f"points following y = {a}x^2 + {b}x + {c} + error")
plt.plot(x, y_hat, label=fr"ML estimator: $\theta_2={theta_ML[2]:.3f}$, $\theta_1={theta_ML[1]:.3f}$, $\theta_0={theta_ML[0]:.3f}$")
plt.legend()
plt.show()
```

## Higher order polynomial

```python
a = -1
b = 9
c = 100
d = -20
e = 4
perturbation = 4000
x = np.linspace(-10, 10)
noise = np.random.random(x.shape) - 0.5
y = a*x**4 + b*x**3 + c*x**2 + d*x + e + noise * perturbation

X = np.array([np.ones_like(x), x, x**2, x**3, x**4]).T

theta_ML = inv(X.T @ X) @ X.T @ y

y_hat = theta_ML @ X.T

plt.scatter(x, y)
plt.plot(x, y_hat)
plt.show()
```

## Sinusoidal

```python
b = 2
perturbation = 1
x = np.linspace(-10, 10)
noise = np.random.random(x.shape) - 0.5
y = np.sin(x) + b + noise * perturbation

X = np.array([np.ones_like(x), np.sin(x)]).T

theta_ML = inv(X.T @ X) @ X.T @ y

y_hat = theta_ML @ X.T

plt.scatter(x, y)
plt.plot(x, y_hat)
plt.show()
```
